# Image Processing

### Introduction
Image processing is a technique to process images to extract some form of information from them. Digital images are processed as a matrix of dimensions mxn which are the resolution of the image. The image is treated as a matrix which holds intensity value for each pixel. The intensity values can either be a single number (for grayscale or single coloured image) or RGB/CMYK value of the colour of that pixel.

### Steps to extract information
#### a) Importing the image
This step involves importing the image and getting it arranged as an mxn array to apply the necessary manipulations.
#### b) Analysing the image
We need to extract data in the form of sex of (x, y) coordinates. We can do this by applying various transforms like edge detection, colour filters, masks etc.
#### c) Using the output
The output generated can be used in whichever way required. For an instance the data could be fed into a machine learning model or used as a dataset to train some ai. Alternatively the extracted data can be plotted and the visual results can be utilised in some studies.

The following code snippet demonstrates how an image is imported as a matrix:

```python
import numpy as np # required to quantise the image as a mxn array
from matplotlib import pyplot as plt # required to plot the final output
import cv2

image = cv2.imread("images/sample.png")
plot = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(plot.shape)
print(type(plot))
# uncomment the next line to see array representation
# print(image)
plt.imshow(plot)
```

### Image splitting into color channels
We can split the image into its individual rgb channels. We can also apply masks and change the image from rgb to grayscale. By default openCV imports images/video streams as BGR instead of the conventional RGB. So we convert the BGR image to RGB before performing any further manipulation. The colorbar on the right of the image represents the intensity of each value.

```python
img = cv2.imread("images/sample1.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
red = img[:, :, 0]
green = img[:, :, 1]
blue = img[:, :, 2]

# uncomment the channel you wish to see
# plt.imshow(red)
# plt.imshow(green)
plt.imshow(blue)

plt.colorbar()
```

### Blurs
You can blur an image using the cv2.blur(image, kernel). It averages out the color in submatrices of dimensions equal to that of the kernel. For the demonstration let's take the image sample.png and apply some blur to it. There are other kinds of blurs as well like Gaussian blur, median blur, bilateral filter. They all have their different usecases and can be used interchangeably depending on the desired result.

```python
blur = cv2.blur(image, (10,10))
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
```
