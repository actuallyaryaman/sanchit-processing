# Image Processing

### Introduction
Image processing is a technique to process images to extract some form of information from them. Digital images are processed as a matrix of dimensions mxn which are the resolution of the image. The image is treated as a matrix which holds intensity value for each pixel. The intensity values can either be a single number (for grayscale or single coloured image) or RGB/CMYK value of the colour of that pixel.

### Steps to extract information
#### a) Importing the image
This step involves importing the image and getting it arranged as an mxn array to apply the necessary manipulations.
#### b) Analysing the image
We need to extract data in the form of sex of (x, y) coordinates. We can do this by applying various transforms like edge detection, colour filters, masks etc.
#### c) Using the output
The output generated can be used in whichever way required. For an instance the data could be fed into a machine learning model or used as a dataset to train some ai. Alternatively the extracted data can be plotted and the visual results can be utilised in some studies.

The following code snippet demonstrates how an image is imported as a matrix:

```python
import numpy as np # required to quantise the image as a mxn array
from matplotlib import pyplot as plt # required to plot the final output
import cv2

image = cv2.imread("images/sample.png")
cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(image.shape)
print(type(image))
plt.imshow(image)
```
